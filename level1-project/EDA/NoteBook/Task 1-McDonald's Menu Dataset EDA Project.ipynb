###  Visualize distribution of 'Calories'
- to identify any skewed data or outliers

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Calories'], bins=20, kde=True,color='Yellow')
plt.title('Calories Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

###  Box plot for Total Fat across different categories

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Total Fat', palette='Set1')
plt.title('Box Plot of Total Fat by Category')
plt.xlabel('Category')
plt.ylabel('Total Fat (g)')
plt.xticks(rotation=45)
plt.show()
```

### Box plot for Sodium across different categories 

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Sodium', palette='Set1')
plt.title('Box Plot of Sodium by Category')
plt.xlabel('Category')
plt.ylabel('Sodium (mg)')
plt.xticks(rotation=45)
plt.show()
```

## 4. EDA

 ### Category Analysis: Average Calories per Category

```python
plt.figure(figsize=(10, 6))
category_avg_calories = df.groupby('Category')['Calories'].mean().sort_values()
category_avg_calories.plot(kind='barh', color='Purple')
plt.title('Average Calories per Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.show()
```

#### Insights:
1. `Non-veg` has High Calories followed by `Smoothies & Shakes` and `Breakfast`.
2. `Beverages` has Low Claories

## Nutritional Analysis - Fat and Carbohydrates

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Total Fat', y='Carbohydrates', hue='Category')
plt.title('Total Fat vs Carbohydrates by Category')
plt.xlabel('Total Fat (g)')
plt.ylabel('Carbohydrates (g)')
plt.legend(loc='upper right')
plt.show()
```

#### Insights:
1. `Smoothies & Shakes` have High Carbohydrates.
2. `Breakfast` includes High Fat and Carbohydrates.
3. `Salads` have Low Fat and Carbohydrates.

### Correlation Heatmap

```python
# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix on the numeric data
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap Matrix')
plt.show()
```

## Conclusion:
The exploratory data analysis (EDA) of McDonald's menu dataset provided valuable insights into the nutritional composition and calorie distribution of its offerings. Key findings include:

- 1.Caloric Content: High-calorie items are prevalent, especially in categories like burgers and desserts, emphasizing the importance of portion control.
- 2.Nutritional Imbalance: Many items exceed daily recommended levels of sodium, sugar, and fat, suggesting limited options for health-conscious consumers.
- 3.Category Comparison: Beverages and sides often have lower caloric density compared to main courses.
- 4.Healthy Options: Salads and lighter meals cater to health-focused consumers but make up a small portion of the menu.

These insights can help McDonald's target menu optimization and personalized marketing strategies to align with consumer preferences and health trends. 

## Recommendations based on EDA:

1. Consider offering lower-calorie options in high-calorie categories.
2. Reduce items with high total and saturated fat to attract health-conscious customers.
3. Promote items with high protein and dietary fiber, which appeal to fitness-oriented customers.
4. Monitor nutritional values, especially for sodium and cholesterol, as excessive values may deter health-focused customers.
